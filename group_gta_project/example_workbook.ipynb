## Import BindingDB data

```python
data = 'data/BindingDB_BindingDB_Articles_202310.tsv'
binding_db = pd.read_csv(data, sep='\t', on_bad_lines='skip', low_memory=False)
binding_db
```

## Data exploration

```python
binding_db.count()
```

### Subset by columns of interest

```python
bio_cols = [
    'Ligand SMILES',
    'Target Name',
    'Ki (nM)',
    'IC50 (nM)',
    'Kd (nM)',
    'EC50 (nM)',
    'Article DOI',
]

bio_df = binding_db[bio_cols]
bio_df.count().plot.bar()
```

### Example: subset by `IC50`

```python
ic50_cols = [
    'Ligand SMILES',
    'Target Name',
    'IC50 (nM)',
]

ic50_df = bio_df[ic50_cols].dropna()
ic50_df
```

### Example: further subset by `Target Name`

```python
ic50_df['Target Name'].value_counts().head()
```

```python
covid19_df = ic50_df[ic50_df['Target Name'] == 'Replicase polyprotein 1ab']
covid19_df = covid19_df.drop_duplicates(subset='Ligand SMILES')
covid19_df
```

## Load in RDKit descriptors

```python
rdkit_desc = pd.read_csv('data/rdkit_descriptors.csv', index_col=0)
rdkit_desc = rdkit_desc.drop_duplicates(subset='SMILES')
rdkit_desc
```

### Slice RDKit descriptors by example subset

```python
covid19_rdkit = rdkit_desc[rdkit_desc['SMILES'].isin(covid19_df['Ligand SMILES'])]
covid19_rdkit
```
